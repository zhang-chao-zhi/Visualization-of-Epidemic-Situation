```python
import matplotlib.pyplot as plt
plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
plt.rcdefaults()
fig, ax = plt.subplots()
# Example data
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
ax.barh(y_pos, performance, xerr=error, align='center',
color='green', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis() # labels read top-to-bottom
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')
plt.show() 
```

```python
class Dp1(object):                                     #
    def __init__(self,n):                              #
        self.mark = [0 for _ in range(n+1)]           #0
        print(self.dp(n))                               #
    def dp(self,n):                                    #
        self.m = 0                                     #mnm
        if self.mark[n] != 0:                          #nmark[n]0,nmark[n]
            self.m = self.mark[n]                      #n
        elif n <= 0:                                   #
            if n == 0:                                 #
                self.m = 1                             #m1,
            else:                                      #
                self.m = 0                             #m0,
        elif n>0:                                      #
            self.m = self.dp(n-2)+self.dp(n-1)         #n  n-1 + n-2
        self.mark[n] = self.m                          #mn
        return self.m                                  #
 
if __name__ == '__main__':
    dp1 = Dp1(100)
```
