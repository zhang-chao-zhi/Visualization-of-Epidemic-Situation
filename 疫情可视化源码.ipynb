```python
import pandas as pd
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Bar, Page, Pie, Timeline,Line,Map
```

```python
data = pd.read_excel('最新数据.xlsx')
```

```python
date = data['日期'].to_list()
```

```python
name = [column for column in data] 
name.pop(0)
```

```python
c = []

for i in range(len(date)):
    
    a1 = data.loc[i:i].values.tolist()
    a1 = a1[0]
    a1.pop(0)
    c.append(a1)
```

```python
for i in range(len(c)):
    c[i].pop(0)
```

```python
def hs(c,k):
    return c[k]
```

```python
#########动态饼图
```

```python
def timeline_pie() -> Timeline:
    attr = name
    tl = Timeline()
    tl.add_schema(is_auto_play = True,
    play_interval = 500,
    is_loop_play = True)
    k= 0
    for i in date:
        pie = (
            Pie()
            .add(
                "",
                [list(z) for z in zip(attr,  hs(c,k))],
            )
                    .set_global_opts(
            title_opts=opts.TitleOpts(title="{}国外疫情数据".format(i)),
            legend_opts=opts.LegendOpts(
                orient="vertical", pos_top="15%", pos_left="85%"
            ),
        )
        )
        tl.add(pie, "{}".format(i))
        k = k + 1
    return tl
timeline_pie().render_notebook()
```
